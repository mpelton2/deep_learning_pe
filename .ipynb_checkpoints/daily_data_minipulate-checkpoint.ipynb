```python
import pandas as pd
import numpy as np
import math
import time
import datetime
```

# Import Indicator Data

```python
df_ind = pd.read_csv('combined_indicators.csv')
df_ind.head()
```

```python
#remove unnamed column
df_indt = df_ind.drop(columns = 'Unnamed: 0')
```

```python
#columns
df_indt.columns = ['Date', 'GDP', 'FEDFUNDS', 'LIBOR', 'MORTGAGE30US', 'REAL_USER_COST',
       'UNEMPLOYMENT_RT']
```

```python
df_indt.columns  
```

```python
#date name change
df_indt['Date'] = df_indt['Date'].apply(lambda x: str(x))
```

```python
#date check
df_indt.iloc[0,0]
```

# Next data

```python
df_sec = pd.read_csv('raw-data.csv')
```

```python
df_sec_2 = df_sec.drop(columns = ['Unnamed: 0'])
```

```python
df_sec_2
```

```python
#date check
df_sec_2.iloc[0,0]
```

# Move returns to quarterly basis

# Combined quarterly return dataframe with quarterly indicator data

```python
joined_raw = df_indt.merge(df_sec_2, on=['Date'], how='outer')
```

## Sort by Date ascending

```python
joined_raw_sort = joined_raw.sort_values(by='Date', ascending=True)
```

```python
joined_raw_sort.iloc[:20,:]
```

# Forward fill all macroeconomic quarterly data

```python
#copy
joined_ffil = joined_raw_sort
```

```python
joined_ffil.columns
```

```python
joined_ffil['GDP'] = joined_ffil['GDP'].ffill()
joined_ffil['FEDFUNDS'] = joined_ffil['FEDFUNDS'].ffill()
joined_ffil['LIBOR'] = joined_ffil['LIBOR'].ffill()
joined_ffil['MORTGAGE30US'] = joined_ffil['MORTGAGE30US'].ffill()
joined_ffil['REAL_USER_COST'] = joined_ffil['REAL_USER_COST'].ffill()
joined_ffil['UNEMPLOYMENT_RT'] = joined_ffil['UNEMPLOYMENT_RT'].ffill()


```

```python
joined_ffil['XLB'] = joined_ffil['XLB'].ffill()
joined_ffil['XLE'] = joined_ffil['XLE'].ffill()
joined_ffil['XLF'] = joined_ffil['XLF'].ffill()
joined_ffil['XLI'] = joined_ffil['XLI'].ffill()
joined_ffil['XLK'] = joined_ffil['XLK'].ffill()
joined_ffil['XLU'] = joined_ffil['XLU'].ffill()
joined_ffil['XLV'] = joined_ffil['XLV'].ffill()
joined_ffil['XLY'] = joined_ffil['XLY'].ffill()
joined_ffil['GSPC'] = joined_ffil['GSPC'].ffill()
```

```python
joined_ffil.tail()
```

# Remove all non market days (when quarterly reports come out on the weekend)

```python
joined_ffil_trim = joined_ffil.dropna()
```

```python
joined_ffil_trim.head()
```

```python
joined_2 = joined_ffil_trim[joined_ffil_trim['LIBOR'] != '.']
```

# Save

```python
joined_2.to_csv('daily_final.csv')
```
